```python
#function to transcribe longer audio files
def voice_recognition(filename):
    model = Model(model_name="vosk-model-en-us-0.22")
    #Model(model_name="vosk-model-en-us-0.15")

    rec = KaldiRecognizer(model, FRAME_RATE)
    rec.SetWords(True)

    mp3 = AudioSegment.from_mp3(filename)
    mp3 = mp3.set_channels(CHANNELS)
    mp3 = mp3.set_frame_rate(FRAME_RATE)

    step = 45000 #vosk doesn't work well with files more than 45secs
    transcript = ""
    for i in range(0,len(mp3),step):
        print(f"Progress: {i/len(mp3)}")
        segment = mp3[i:(i+step)]

        rec.AcceptWaveform(segment.raw_data)
        result = rec.Result()

        text = json.loads(result)["text"]
        transcript += text

    cased = subprocess.check_output("python3 /Users/user/Desktop/Speech_Recognition_and_Summarization_system/recasepunc/recasepunc.py predict /Users/user/Desktop/Speech_Recognition_and_Summarization_system/recasepunc/checkpoint", shell=True, text=True, input=transcript)
    return cased
```

```python
#building a summarizer that can automatically summarize the audio file
from transformers import pipeline

summarizer = pipeline("summarization")
```

```python
with open("transcript.txt") as f:
    transcript = f.read()
```

```python
split_tokens = transcript.split(" ")
docs = []

for i in range(0, len(split_tokens), 850):
    selection = " ".join(split_tokens[i:(i+850)])
    docs.append(selection)
```

```python
docs
```

```python
summaries = summarizer(docs)
```

```python
summary = "\n\n".join(d["summary_text"] for d in summaries)
```
